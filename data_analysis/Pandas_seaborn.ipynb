```python
import pandas as pd
import seaborn
%matplotlib notebook

iris = pd.read_csv('C:/Selvaria/DataAnalysis/iris.csv')
#seaborn.barplot(x = 'species',y = 'sepal_length',data = iris,ci=90,n_boot = 1000)#柱状图
seaborn.violinplot(x = 'species',y = 'sepal_length',data = iris,ci=90,n_boot = 1000)
```

```python
import math
l = [3,5,9,12,7,8,6,8]

avg = sum(l)/len(l)
sum1 =0
for i in l:
    mid = (i-avg)**2
    sum1 = sum1+mid
sd = math.sqrt(sum1/(len(l)-1))
print(sd)
```

```python
#seaborn.distplot(iris['petal_width'])#直方图
seaborn.countplot('petal_width',data=iris)#条形图，计数类名出现的次数
```

```python
#seaborn.pointplot(x='species',y='sepal_width',data=iris)#折线图
seaborn.regplot(x='sepal_length',y='sepal_width',data = iris)#散点图
seaborn.regplot(x='petal_length',y='petal_width',data = iris)
```

```python
seaborn.boxplot(x = 'species',y='petal_width',data = iris)#箱型图
```

```python
iris_vir = iris[iris.species=='Iris-virginica']
iris_s = iris[iris.species=='Iris-setosa']
iris_ver = iris[iris.species=='Iris-versicolor']

seaborn.distplot(iris_vir['petal_width']).set(ylim=(0,12))#.set(ylim=(0,12))为限制y轴范围
seaborn.distplot(iris_s['petal_width'])
seaborn.distplot(iris_ver['petal_width'],label='vir').legend()#说明图形名称
```

```python
q =seaborn.FacetGrid(iris,hue='species')#hue是同时画在一个图里，row是一行分别画三个，col是一列画三个
q.set(xlim=(0,8))
q.map(seaborn.regplot,'petal_length','petal_width').add_legend()
```

```python
import matplotlib.pyplot as plt
q =seaborn.FacetGrid(iris,hue='species')
q.set(xlim=(0,2.5))
q.map(plt.scatter,'petal_width','petal_length').add_legend()
```
