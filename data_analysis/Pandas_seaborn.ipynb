```python
import pandas as pd
import seaborn
%matplotlib notebook

iris = pd.read_csv('C:/Selvaria/DataAnalysis/iris.csv')
#seaborn.barplot(x = 'species',y = 'sepal_length',data = iris,ci=90,n_boot = 1000)#æŸ±çŠ¶å›¾
seaborn.violinplot(x = 'species',y = 'sepal_length',data = iris,ci=90,n_boot = 1000)
```

```python
import math
l = [3,5,9,12,7,8,6,8]

avg = sum(l)/len(l)
sum1 =0
for i in l:
    mid = (i-avg)**2
    sum1 = sum1+mid
sd = math.sqrt(sum1/(len(l)-1))
print(sd)
```

```python
iris_vir = iris[iris.species=='Iris-virginica']
iris_s = iris[iris.species=='Iris-setosa']
iris_ver = iris[iris.species=='Iris-versicolor']

seaborn.distplot(iris_vir['petal_width']).set(ylim=(0,12))#.set(ylim=(0,12))ä¸ºé™åˆ¶yè½´èŒƒå›´
seaborn.distplot(iris_s['petal_width'])
seaborn.distplot(iris_ver['petal_width'],label='vir').legend()#è¯´æ˜Žå›¾å½¢åç§°
```
